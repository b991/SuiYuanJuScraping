```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SYJdf = pd.read_csv("./SuiYuanJuResults.csv")
```

Process date and time so that combined_datetime is a datetime object



```python
SYJdf['combined_datetime'] = SYJdf['date'] + ' ' +SYJdf['time']
SYJdf['combined_datetime']=pd.to_datetime(SYJdf['combined_datetime'])
```

Convert UTC to Beijing Standard Time. Seperate converted date from converted time.

It's important to make converted_time into int, otherwise the plot won't come out correctly.

```python
SYJdf['converted_date'] = SYJdf['combined_datetime'].dt.tz_localize('UTC').dt.tz_convert('Asia/Shanghai').dt.strftime('%Y-%m-%d')
SYJdf['converted_time'] = SYJdf['combined_datetime'].dt.tz_localize('UTC').dt.tz_convert('Asia/Shanghai').dt.strftime('%H').astype(int)
```
